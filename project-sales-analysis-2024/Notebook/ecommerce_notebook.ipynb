```python
# Copier le DataFrame original
df_filled = df_or.copy()

# Étape 1 : Créer un dictionnaire de référence par StockCode
stockcode_ref = df_filled.dropna(subset=['Description', 'UnitPrice']) \
                         .groupby('StockCode') \
                         .agg({'Description': 'first', 'UnitPrice': 'first'})

df_filled['Description'] = df_filled.apply(
    lambda row: stockcode_ref.loc[row['StockCode'], 'Description']
    if pd.isna(row['Description']) and row['StockCode'] in stockcode_ref.index
    else row['Description'],
    axis=1
)

df_filled['UnitPrice'] = df_filled.apply(
    lambda row: stockcode_ref.loc[row['StockCode'], 'UnitPrice']
    if pd.isna(row['UnitPrice']) and row['StockCode'] in stockcode_ref.index
    else row['UnitPrice'],
    axis=1
)

```
